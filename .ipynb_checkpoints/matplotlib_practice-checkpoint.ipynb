```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Matlab Method
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label="Boston")

# Create labels for the x and y axis
plt.xlabel('Date')
plt.ylabel("Fare ($)")

# Set the y limit between - and 45.
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()

# Add the legend
plt.legend()

```

```python
#Scatter Plot Using MatLab Method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create Plot
plt.scatter(x_axis, y_axis, color = "r", label = "Chicago")
plt.gca().invert_yaxis()

# Create labels
plt.xlabel("Fare )$)")
plt.ylabel("Date")

# Create title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend("Chicago")
```

```python
#Bubble Using MatLab Method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Multiply y_axis to increase buble size
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
# Create Plot
plt.scatter(x_axis, y_axis, color = "r", label = "Chicago", s = y_axis_larger)
plt.gca().invert_yaxis()

# Create labels
plt.xlabel("Fare )$)")
plt.ylabel("Date")

# Create title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend("Chicago")
```

```python
# Matplotlib Object Oriented  Method  Scatter Chart
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot and add a label for the legend
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "xkcd:sky blue", s=[i*5 for i in y_axis], alpha=0.5, edgecolor = "b", linewidths = 2)
ax.invert_yaxis()
ax.set_xlim(0, 50)

# Create labels for the x and y axis
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")


# Create a title
ax.set_title("PyBer Fare by Month")


# Add the legend
ax.legend("Boston")
```

```python
#Pie Chart Using MatLab Method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Set colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred"]

#Explode 1 item
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
# Create Plot
plt.subplots(figsize = (8, 8))
plt.pie(y_axis,
        labels = x_axis,
        colors=colors,
        explode=explode_values,
        autopct="%.1f%%")


plt.show()
```

```python
# Matplotlib Object Oriented  Method
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]



colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "grey", "brown", "tan", "pink"]
#Explode 1 item
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)

# Create the plot and add a label for the legend
fig, ax = plt.subplots(figsize= (8, 8))
ax.pie(y_axis,
       labels = x_axis,
       explode=explode_values,
       autopct="%.1f%%",
      shadow = "True",
      startangle = 85,
      colors = colors)

# Invert order
ax.invert_xaxis()

# Create a title
ax.set_title("PyBer Fare by Month")


plt.show()
```
